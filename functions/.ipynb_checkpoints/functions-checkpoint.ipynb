# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

Пользователь по команде "p" может узнать владельца документа по его номеру

````python
def command_p():
    '''
    Вывод на консоль значения словаря по ключу, если он есть в списке словарей
    '''
    user_input = input('Введите номер документа: ')
    if_found = 0
    for line in documents:
        if line['number'] == user_input:
            print(f"Владелец документа:  {line['name']}")
            if_found += 1
    if if_found == 0:
        print('Документ не найден в базе')
````

Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

````python
def command_s():
    '''
    Вывод на консоль ключа словаря по значению, если он есть в списке словарей
    '''
    user_input = input('Введите номер документа: ')
    if_found = 0
    for key, value in directories.items():
        if user_input in value:
            print(f'Документ хранится на полке: {key}')
            if_found += 1
    if if_found == 0:
        print('Документ не найден в базе')
````

Пользователь по команде "l" может увидеть полную информацию по всем документам

````python
def command_l():
    '''
    Вывод на консоль элементов списка и словаря в требуемом формате.
    '''
    print('Текущий список документов: ')
    for line in documents:
        print(f"№: {line['number']}, тип: {line['type']}, владелец: {line['name']}, ", end='')
        for key, value in directories.items():
            if line['number'] in value:
                print(f'полка хранения: {key}')
````

Пользователь по команде "ads" может добавить новую полку

````python
def print_keys_from_directories():
    '''
    Вывод на печать ключей словаря directories в требуемом виде
    '''
    print(f"Текущий перечень полок: ", end='')
    for g in list(directories.keys()):
        if g != list(directories.keys())[-1]:
            print(f"{g}, ", end='')
        else:
            print(f"{g}. ")
````

````python
def command_ads():
    '''
    Добавляет новый ключ в словарь если его нет
    Параметр - ввод пользователя
    '''
    user_input = input('Введите номер полки: ')
    if user_input not in directories:
        directories[user_input] = []
        print(f"Полка добавлена. ", end='')
        print_keys_from_directories()
    else:
        print('Такая полка уже существует. ', end='')
        print_keys_from_directories()
````

Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

````python
def command_ds():
    '''
    Удаляет ключ из словаря, если значение отсутствует
    Параметр - ввод пользователя
    '''
    user_input = input('Введите номер полки: ')
    if user_input not in directories:
        print('Такой полки не существует. ', end='')
        print_keys_from_directories()
    elif user_input in directories and directories[user_input] != []:
        print('На полке есть документы, удалите их перед удалением полки.')
        print_keys_from_directories()
    else:
        del(directories[user_input])
        print(f"Полка удалена. ", end='')
        print_keys_from_directories()
````

````python
def main():
    commands_list = {
        'p': command_p,
        's': command_s,
        'l':command_l,
        'ads':command_ads,
        'ds':command_ds
    }
    while True:
        command = input('Введите команду: ')
        if command in commands_list:
            res = commands_list[command]()
        else:
            print('Вы хотите чего-то странного. Подумайте еще раз)')
        exit = input('Для выхода из программы введите q, для продолжения нажмите любую клавишу ').lower()
        if exit == 'q':
            break
````

````python
main()
````

````python
def command_ad():
    '''
    Добавляет новые элементы в список и словарь.
    '''
    user_input_num = input('Введите номер документа: ')
    user_input_type = input('Введите тип документа: ')
    user_input_name = input('Введите владельца документа: ')
    user_input_shelf = input('Введите полку для хранения: ')
    if user_input_shelf not in directories:
        print('Такой полки не существует. Добавьте полку командой as.')
        command_l()
    else:
        documents.append({'type': user_input_type, 'number': user_input_num, 'name': user_input_name})
        directories[user_input_shelf].append(user_input_num)
        print('Документ добавлен.')
        command_l()
````

````python
def command_d():
    '''
    Удаляет элементы из списка и словаря.
    '''    
    user_input_num = input('Введите номер документа: ')
    documents_copy = documents.copy()
    if_found = 0
    for line in documents_copy:
        if line['number'] == user_input_num:
            documents.remove(line)
            for key, value in directories.items():
                if line['number'] in value:
                    value.remove(line['number'])
            print('Документ удален.')
            command_l()
            if_found += 1
    if if_found == 0:
        print('Документ не найден в базе')
        command_l()
````

````python
def command_m():
    '''
    При совпадении ввода пользователя и ключа и одного из его значений 
    копирует с удалением из предедущего значения ключа в значение ключа, введенного пользователем
    '''
    user_input_num = input('Введите номер документа: ')
    user_input_shelf = input('Введите номер полки: ')
    if_found = 0         
    if user_input_shelf not in directories:
        print('Такой полки не существует')
        print_keys_from_directories()
    for value in directories.values():      
        if user_input_num in value:
            value.remove(user_input_num)
            directories[user_input_shelf].append(user_input_num)
            print('Документ перемещен')
            command_l()
            if_found += 1
            break
    if if_found == 0:
        print('Документ не найден в базе')
        command_l()
````

````python
def main():
    commands_list = {
        'p': command_p,
        's': command_s,
        'l':command_l,
        'ads':command_ads,
        'ds':command_ds,
        'ad':command_ad,
        'd':command_d,
        'm':command_m
    }
    while True:
        command = input('Введите команду: ')
        if command in commands_list:
            res = commands_list[command]()
        else:
            print('Вы хотите чего-то странного. Подумайте еще раз)')
        exit = input('Для выхода из программы введите q, для продолжения нажмите любую клавишу ').lower()
        if exit == 'q':
            break
````

````python
main()
````
